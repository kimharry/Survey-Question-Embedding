```python
import pandas as pd

data = pd.read_csv('data/na-removed-data-v3.csv')
```

```python
# replace "Less than 1 year" from YearsCodePro with 0.5
data["YearsCodePro"] = data["YearsCodePro"].replace("Less than 1 year", 0.5)
data["YearsCodePro"] = pd.to_numeric(data["YearsCodePro"])

data.to_csv("data/na-removed-data-v4.csv", index=False)
```

```python
type(data["CodingActivities"].values.tolist())
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')

embedding_columns = [
    "CodingActivities",
    "LearnCode",
    "LearnCodeOnline",
    "BuyNewTool",
    "TechEndorse",
    "LanguageHaveWorkedWith",
    "LanguageWantToWorkWith",
    "LanguageAdmired",
    "DatabaseHaveWorkedWith",
    "DatabaseWantToWorkWith",
    "DatabaseAdmired",
    "PlatformHaveWorkedWith",
    "PlatformWantToWorkWith",
    "PlatformAdmired",
    "WebframeHaveWorkedWith",
    "WebframeWantToWorkWith",
    "WebframeAdmired",
    "EmbeddedHaveWorkedWith",
    "EmbeddedWantToWorkWith",
    "EmbeddedAdmired",
    "MiscTechHaveWorkedWith",
    "MiscTechWantToWorkWith",
    "MiscTechAdmired",
    "ToolsTechHaveWorkedWith",
    "ToolsTechWantToWorkWith",
    "ToolsTechAdmired",
    "NEWCollabToolsHaveWorkedWith",
    "NEWCollabToolsWantToWorkWith",
    "NEWCollabToolsAdmired",
    "OpSysPersonal use",
    "OpSysProfessional use",
    "OfficeStackAsyncHaveWorkedWith",
    "OfficeStackAsyncWantToWorkWith",
    "OfficeStackAsyncAdmired",
    "OfficeStackSyncHaveWorkedWith",
    "OfficeStackSyncWantToWorkWith",
    "OfficeStackSyncAdmired",
    "AISearchDevHaveWorkedWith",
    "AISearchDevWantToWorkWith",
    "AISearchDevAdmired",
    "AIBen",
    "AIToolCurrently Using",
    "AIToolInterested in Using",
    "AIToolNot interested in Using",
    "AINextMuch more integrated",
    "AIEthics",
    "AIChallenges",
    "Frustration",
    "ProfessionalTech"
]

new_data = pd.DataFrame()

for column in data.columns:
    if column in embedding_columns:
        data[column] = data[column].fillna('')
        try:
            new_data[column] = model.encode(data[column]).tolist()
        except:
            print("emb", column)
    else:
        # if all values are numeric, pass
        # else, change to numbering with the order of the number of each unique value
        if data[column].dtype == 'int64' or data[column].dtype == 'float64':
            new_data[column] = data[column]
        else:
            data[column] = data[column].fillna('')
            uniques = data[column].unique()
            try:
                uniques.sort()
            except:
                print("sort", column)
            temp_column = data[column].copy()
            for i, unique in enumerate(uniques):
                temp_column.loc[temp_column == unique] = i
            new_data[column] = temp_column
```
